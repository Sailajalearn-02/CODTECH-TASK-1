```python
k=['UDI', 'Product ID', 'Type', 'Air temperature [K]',
        'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]',
        'Tool wear [min]', 'Target', 'Failure Type']
for i in k:
  j=df[i].value_counts()
  print(j)
```

# correlation Analysis


```python
cor=df.corr(numeric_only=True)
cor
```

# Data Visualization
# Heat map

```python
sns.heatmap(cor,xticklabels=cor.columns,yticklabels=cor.columns,annot=True,cmap="Reds",fmt=".2f")
```

# Box Plot

```python
for i in df.columns:
  plt.figure(figsize=(10,7))
  sns.boxplot(df[i])
  plt.show()
```

# Dist Plot

```python
for i in df.columns:
  if df[i].dtype != object:
    sns.displot(df[i],kde=True,color='blue')
    plt.show()
```

# scatter Plot

```python
sns.scatterplot(x='Failure Type',y='Rotational speed [rpm]',data=df,hue='Type',palette='rocket')
plt.xticks(rotation=90)
plt.title('failure Type vs Rotational speed[rpm]')
plt.show()
```

```python
sns.scatterplot(x='Failure Type',y='UDI',data=df,hue='Type',palette='rocket')
plt.xticks(rotation=90)
plt.title(' Failure Type vs UDI')
plt.show()
```

```python
sns.scatterplot(x='Failure Type',y='Air temperature [K]',data=df,hue='Type',palette='rocket')
plt.xticks(rotation=90)
plt.title(' Failure Type vs Air Temperature')
plt.show()
```

```python
sns.scatterplot(x='Failure Type',y='Torque [Nm]',data=df,hue='Type',palette='rocket')
plt.xticks(rotation=90)
plt.title(' Failure Type vs Torque [Nm]')
plt.show()
```

```python
sns.scatterplot(x='Failure Type',y='Tool wear [min]',data=df,hue='Type',palette='rocket')
plt.xticks(rotation=90)
plt.title(' Failure Type vs Tool wear [min]')
plt.show()
```

```python
sns.scatterplot(x='Failure Type',y='Target',data=df,hue='Type',palette='rocket')
plt.xticks(rotation=90)
plt.title(' Failure Type vs Target')
plt.show()
```
